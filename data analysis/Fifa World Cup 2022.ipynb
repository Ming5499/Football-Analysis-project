TOTAL NUMBER OF TEAMS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
wc = pd.read_csv('data/FIFA_World_Cup_2022_players.csv')
```

Total number of nations

```python
wc['Nation'].nunique()
```

Position

```python
position_distribution = wc['Position'].value_counts(normalize=True) * 100
plt.figure(figsize=(8, 8))
plt.pie(position_distribution, labels=position_distribution.index,shadow = True, autopct='%1.1f%%', startangle=140)
plt.title('Position Distribution in All Teams %', y = 1.2, weight = 'bold')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

Club

```python

plt.figure(figsize = (12,12))
xy = sns.barplot(y = wc['Club'].value_counts().head(10).index, x = wc['Club'].value_counts().head(10).values)
for i in range(10):
    value = wc['Club'].value_counts().head(10)[i]
    xy.text(x = value - 0.5  ,y = i + 0.1 , s = value, color='white', ha="center", fontsize = 17)
    
plt.title('Clubs with the most players', weight = 'bold')
plt.yticks(fontsize = 13)
plt.show()
```

Average age of players for each nation

```python
average_age_by_nation = wc.groupby('Nation')['Age'].mean()
average_age_by_nation
```

```python
plt.figure(figsize=(12, 6))
average_age_by_nation.plot(kind='bar', color='skyblue')

plt.title('Average Age of Players by Nation')
plt.xlabel('Nation')
plt.ylabel('Average Age')
plt.xticks(rotation=45, ha='right')  #
plt.grid(axis='y')  
plt.tight_layout()

plt.show()
```
